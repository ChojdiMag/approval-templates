<h2>Laden vehicles and trailers</h2>
<p>The relationships required by the diagram shall apply progressively for intermediate states of loading 
between the laden and the unladen states and shall be achieved by automatic means.</p>


```python
x = [100,450,750]
```

```python
y=[0,0.35,0.575]
```

```python
x1=[20,750]
```

```python
y1=[0,0.80]
```

```python
fig = plt.plot(x,y),plt.plot(x1,y1)
```

<p>Brake results - laden</p>
<p>Fill pressure of break(kPa) and braking efficiency(E_per) from Vbox</p>

```python
kPa=[100,200,300,400,500,600,650]
```

```python
E_per =[0.04,0.1,0.4,0.36,0.42,0.49,0.5]
```

```python
plt.figure(figsize=(7,8))
plt.title('Braking efficiency at pressure - laden vehicle')
fig = plt.plot(x,y,label='lower limit',color='green'),plt.plot(x1,y1,label='upper limit',color='green'),plt.plot(kPa,E_per,label='test result',color='red')
plt.xlabel('kPa')
plt.ylabel('braking factor')
plt.legend()
```
